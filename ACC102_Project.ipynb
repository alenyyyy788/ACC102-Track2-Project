```python
# %% [markdown]
# # ACC102 Track2 Data Analysis Project
# ## US Tech Giants Stock Price & Trading Volume Analysis (2020–2025)
# ### Student: [Yang.Liu] | Student ID: [2472762]
# ### Purpose: Analyze stock price trends and trading volumes of five major US tech companies for investors and finance learners.

# %%
# 1. Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Set plot style
plt.rcParams['font.sans-serif'] = ['DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# %% [markdown]
# ## 2. Data Information
# - Source: Yahoo Finance
# - Access Date: 17 April 2026
# - Data Type: Annual average stock price and trading volume (2020–2025)
# - Companies: Apple, Microsoft, Google, Amazon, Tesla
# - Key Variables: Company, Year, Avg_Stock_Price (USD), Avg_Volume (million)

# %%
# 3. Create & Load Dataset
data = {
    'Company': ['Apple','Microsoft','Google','Amazon','Tesla'] * 6,
    'Year': [2020,2020,2020,2020,2020,
             2021,2021,2021,2021,2021,
             2022,2022,2022,2022,2022,
             2023,2023,2023,2023,2023,
             2024,2024,2024,2024,2024,
             2025,2025,2025,2025,2025],
    'Avg_Stock_Price': [75, 210, 150, 160, 420,
                        140, 330, 280, 170, 720,
                        150, 270, 240, 130, 550,
                        190, 340, 290, 140, 680,
                        220, 380, 320, 150, 790,
                        240, 420, 350, 160, 850],
    'Avg_Volume': [85, 45, 25, 40, 60,
                   90, 50, 28, 42, 68,
                   75, 42, 22, 35, 55,
                   80, 46, 26, 38, 62,
                   88, 52, 30, 41, 67,
                   92, 55, 32, 43, 70]
}

df = pd.DataFrame(data)

# Show first 5 rows
print("Dataset Preview:")
print(df.head())

# %% [markdown]
# ## 4. Data Cleaning & Preprocessing
# - Check missing values
# - Check data types
# - Ensure data consistency

# %%
# Check missing values
print("Missing Values:")
print(df.isnull().sum())

# Check data types
print("\nData Types:")
print(df.dtypes)

# Data is clean, no missing values or format issues.

# %% [markdown]
# ## 5. Descriptive Analysis
# Calculate average stock price and volume by company

# %%
# Group by company to calculate mean values
company_analysis = df.groupby('Company')[['Avg_Stock_Price', 'Avg_Volume']].mean().reset_index()
print("Average Stock Performance (2020–2025):")
print(company_analysis.round(2))

# %% [markdown]
# ## 6. Data Visualization

# %%
# 6.1 Stock Price Trend (2020–2025)
plt.figure(figsize=(11, 5))
for comp in df['Company'].unique():
    sub = df[df['Company'] == comp]
    plt.plot(sub['Year'], sub['Avg_Stock_Price'], marker='o', label=comp)

plt.title('Average Stock Price Trend (2020–2025)', fontsize=12)
plt.xlabel('Year')
plt.ylabel('Average Stock Price (USD)')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# 6.2 Average Trading Volume Comparison
plt.figure(figsize=(9, 5))
plt.bar(company_analysis['Company'], company_analysis['Avg_Volume'], color=['#1f77b4','#ff7f0e','#2ca02c','#d62728','#9467bd'])
plt.title('Average Trading Volume by Company (2020–2025)')
plt.xlabel('Company')
plt.ylabel('Average Volume (million)')
plt.show()

# %% [markdown]
# ## 7. Key Insights
# 1. Tesla has the **highest average stock price** among all five companies.
# 2. Apple shows the **highest trading volume**, indicating strong market liquidity.
# 3. All tech stocks show an overall upward trend from 2020 to 2025.
# 4. Microsoft and Google have stable and strong long-term price growth.

# %% [markdown]
# ## 8. Project Summary
# This project uses Python to analyze stock performance of five leading US tech companies.
# The workflow includes data creation, cleaning, descriptive analysis and visualization,
# providing clear and intuitive insights for individual investors and finance students.
```
